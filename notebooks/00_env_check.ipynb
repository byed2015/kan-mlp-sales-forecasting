```python
import sys, platform, os, subprocess
print("Python:", sys.version)
print("OS:", platform.platform())
print("Working dir:", os.getcwd())

# Versiones clave
import numpy, pandas, sklearn
print("NumPy:", numpy.__version__)
print("Pandas:", pandas.__version__)
print("scikit-learn:", sklearn.__version__)

# Opcionales
try:
    import torch, torchvision, torchaudio
    print("PyTorch:", torch.__version__)
    print("CUDA disponible:", torch.cuda.is_available())
    if torch.cuda.is_available():
        print("Nombre GPU:", torch.cuda.get_device_name(0))
except Exception as e:
    print("PyTorch no disponible o error:", e)

for pkg in ["einops", "optuna", "lightning", "seaborn", "plotly"]:
    try:
        m = __import__(pkg)
        print(f"{pkg}:", m.__version__ if hasattr(m,"__version__") else "OK")
    except Exception as e:
        print(f"{pkg} no disponible:", e)

# pykan
try:
    import kan
    print("KAN (pykan) OK")
except Exception as e:
    print("KAN no disponible (pykan):", e)
```
